```python
#중복제거
import pandas as pd
movie_insta_df = pd.DataFrame([])
folder = './'
f_list = ['insta_love_crawling.csv', 'insta_instagood_crawling.csv','insta_photooftheday_crawling.csv',
          'insta_fashion_crawling.csv','insta_beautiful_crawling.csv','insta_happy_crawling.csv',
         'insta_cute_crawling.csv','insta_tbt_crawling.csv','insta_like4like_crawling.csv','insta_followme_crawling.csv']
for f_name in f_list:
    f_path = folder + f_name
    tmp = pd.read_csv(f_path)
    print(tmp.head())
    movie_insta_df = movie_insta_df.append(tmp)
movie_insta_df.columns = ['content', 'date', 'like', 'place', 'tags']
movie_insta_df.drop_duplicates(subset = ['content'], inplace=True)  # inplace=True 옵션을 통해 원본 데이터 변경 
movie_insta_df.to_csv('./tag_crawling.csv', index = False)
```

```python
#해시태그 수집
import pandas as pd 
raw_total = pd.read_csv('./tag_crawling.csv')
tags_total = []
for tags in raw_total['tags']:
    # print(tags)
    tags_list = tags[2:-2].split("', '") #[' ~ '] 제외
    for tag in tags_list:
        tags_total.append(tag)
#빈도수가 많은 10가지 해시태그 출력 
from collections import Counter
tags_count = Counter(tags_total)
print(tags_count.most_common(10))
```
